```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
import joblib
import pandas as pd
from imblearn.over_sampling import SMOTE
```

# Opening Processed Data

```python
df = pd.read_csv("Processed_Heart_Data.csv")
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

# Fixing Label Unbalance Using Over-Sampling

```python
# Splitting X (Features) and Y (Lable)
x=df.drop('HeartDisease',axis=1).values
y=df['HeartDisease'].values
print("X.Shape: ", x.shape)
print("Y.Shape: ", y.shape)
```

```python
# Applying Over-Sampling
sm = SMOTE(random_state=42)
X, Y = sm.fit_resample(x, y)
```

```python
# Splitting test and train datasets
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
print("X_train.Shape: ", x_train.shape)
print("y_train.Shape: ", y_train.shape)
print("X_test.Shape: ", x_test.shape)
print("y_test.Shape: ", y_test.shape)
```

# Defining the Classifier: XGBoost

```python
# Creating the classifier
xgb = XGBClassifier(n_estimators= 500 , max_depth= 3 , learning_rate = 0.1)
# Training the Classifier on train set
xgb.fit(x_train , y_train);
```

```python
print("Training Report:")
print(classification_report(y_test,xgb.predict(x_test)))
```

```python
print ("Test Accuracy (F1 Score): ", xgb.score(x_test , y_test))
```

# Saving Model for Later User

```python
joblib.dump(xgb, "classifer.model") 
```
